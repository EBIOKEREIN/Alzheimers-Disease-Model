## Logisitic Regression Model

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['DoctorInCharge','PatientID','Diagnosis'], axis=1).values
y = df['Diagnosis'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 123, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

X_train_std = scale.fit_transform(X_train)
X_test_std = scale.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000)

model.fit(X_train_std, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
# training error
accuracy_score(y_train, model.predict(X_train_std))
```

```python
# testing error
accuracy_score(y_test, model.predict(X_test_std))
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, model.predict_proba(X_test_std)[:,1])

plt.plot(thresholds,precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label = 'Recall')
plt.xlabel('Threshold')
plt.ylabel('Precision/Recall Score')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (1)
y_prob = model.predict_proba(X_test)[:, 1]

# Compute the ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(7, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')

plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
tn, fp, fn, tp = cm.to_numpy().ravel() 
tn, fp, fn, tp
```

```python
#=======accuracy, specificity, sensitivity 

tn, fp, fn, tp = cm.to_numpy().ravel()   # unpack confusion matrix

accuracy    = (tp + tn) / (tp + tn + fp + fn)
sensitivity = tp / (tp + fn)   # true positive rate, recall
precision = tp / (tp + fp) # precision
specificity = tn / (tn + fp)   # true negative rate

print(f"Accuracy:    {accuracy:.3f}")
print(f"Recall: {sensitivity:.3f}")
print(f"Specificity: {specificity:.3f}")
print(f"Precision: {precision:.3f}")
```

```python
from re import I
from sklearn.model_selection import train_test_split

X = new_df.drop(columns=['DoctorInCharge','PatientID','Diagnosis'])
y = new_df['Diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Integer
from sklearn.pipeline import Pipeline
import xgboost as xgb

# Create pipeline
pipe = Pipeline([
    ('clf', xgb.XGBClassifier(random_state=8))
])

# Define search space
search_space = {
    'clf__n_estimators': Integer(50, 500),
    'clf__max_depth': Integer(2, 8),
    'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
    'clf__subsample': Real(0.5, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode': Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.0, 10.0),
    'clf__reg_lambda': Real(0.0, 10.0),
    'clf__gamma': Real(0.0, 10.0)
}

# Create BayesSearchCV
opt = BayesSearchCV(
    pipe, 
    search_space, 
    cv=3, 
    n_iter=50,  
    scoring='roc_auc', 
    random_state=42,
    n_jobs=-1  
)

# Fit it
opt.fit(X_train, y_train)

# Get results
print("Best Parameters:", 
opt.best_params_)
print("Best Score:", opt.best_score_)
```

```python
pipe = Pipeline([
    ('clf', XGBClassifier(
        colsample_bytree=0.9589359959508708,
        colsample_bylevel=0.839355765549384,
        colsample_bynode=0.6975430350538748,
        reg_alpha=0.04882957024517021,
        reg_lambda=3.6360588498404796,
        gamma=9.175875975919723,
        n_estimators=69,
        max_depth=3,
        learning_rate=0.4504289623322671,
        subsample=0.5013236450144629,
        random_state=42
    ))
])

pipe.fit(X_train, y_train)
opt = pipe
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score


# Extract the XGBoost model from the pipeline
xgboost_model = opt.named_steps['clf']

# Create feature importance plot with custom formatting
fig, ax = plt.subplots(figsize=(12, 8))
plot_importance(xgboost_model, ax=ax, max_num_features=20, importance_type='gain')

# Format the x-axis labels to 2 decimal places
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.2f}'))

plt.title('XGBoost Feature Importance')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()
```
